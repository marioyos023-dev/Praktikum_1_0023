```python
import matplotlib.pyplot as plt

g = 9.8
h = 100
t = list(range(11))
s = [h - 0.5 * g * ti**2 for ti in t]
v = [g * ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Gerak Jatuh Bebas (Kasus 1)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 1:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")


g = 9.8
h = 120
t = list(range(11))
s = [h - 0.5 * g * ti**2 for ti in t]
v = [g * ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Jatuh Bebas dengan Kecepatan Awal (Kasus 2)")
plt.xlabel(" (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 2:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")


g = 1.6
h = 50
t = list(range(21))
s = [h - 0.5 * g * ti**2 for ti in t]
v = [g * ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Planet Lain (Gravitasi Beda) (Kasus 3)")
plt.xlabel(" (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 3:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")

g = 9.8
h = 500
t = list(range(21))
s = [h - 0.5 * g * ti**2 for ti in t]
v = [g * ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Jatuh Bebas dengan Waktu Lebih Panjang) (Kasus 4)")
plt.xlabel(" (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 4:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")

```
